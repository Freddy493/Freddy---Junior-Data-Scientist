```python
# Este código instala la biblioteca opendatasets, que permite descargar conjuntos de datos directamente desde Kaggle. Luego, importa varias bibliotecas útiles para el análisis de datos y la visualización, como pandas, numpy, matplotlib y seaborn. A continuación, se define un enlace a un conjunto de datos específico en Kaggle sobre la predicción de precios de laptops y se descarga utilizando la función od.download(). Después, cambia el directorio de trabajo al que contiene el conjunto de datos descargado y lista los archivos en ese directorio. Finalmente, carga el archivo CSV "Laptop_price.csv" en un DataFrame de pandas llamado df.
# This code installs the opendatasets library, which allows downloading datasets directly from Kaggle. It then imports several useful libraries for data analysis and visualization, such as pandas, numpy, matplotlib, and seaborn. Next, it defines a link to a specific dataset on Kaggle about laptop price prediction and downloads it using the od.download() function. After that, it changes the working directory to the one containing the downloaded dataset and lists the files in that directory. Finally, it loads the "Laptop_price.csv" file into a pandas DataFrame called df.
!pip install opendatasets
import opendatasets as od
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns

dataset_link = "https://www.kaggle.com/datasets/mrsimple07/laptoppriceprediction"
od.download(dataset_link)
os.chdir("laptoppriceprediction")
os.listdir()

archivo = "Laptop_price.csv"
df = pd.read_csv(archivo)
```

*   Storage_Capacity: Simulates different storage capacities with options of 256GB, 512GB, and 1000GB (1TB).

*   Screen_Size: Represents the size of the laptop screen, randomly generated between 11 and 17 inches.

*   Weight: Indicates the weight of the laptop in kilograms, simulated uniformly between 2.0 and 5.0 kg.

```python
# Este código calcula la matriz de correlación del DataFrame df utilizando el método corr() de Pandas, que mide la relación entre las diferentes variables numéricas en el conjunto de datos. Luego, se crea una figura con un tamaño específico (8x6) y se utiliza la función heatmap de Seaborn para visualizar esta matriz de correlación. El argumento annot=True permite mostrar los valores de correlación en cada celda, mientras que cmap='coolwarm' establece el esquema de colores para el mapa de calor. El formato de los números se especifica como '.2f' para mostrar dos decimales. Finalmente, se añade un título al gráfico y se muestra con plt.show().
# This code calculates the correlation matrix of the DataFrame df using the corr() method from Pandas, which measures the relationship between different numerical variables in the dataset. Then, a figure is created with a specific size (8x6), and the heatmap function from Seaborn is used to visualize this correlation matrix. The argument annot=True allows displaying the correlation values in each cell, while cmap='coolwarm' sets the color scheme for the heatmap. The number format is specified as '.2f' to show two decimal places. Finally, a title is added to the plot and displayed with plt.show().
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matriz de Correlación")
plt.show()
```

Esto resalta la importancia de esta característica en nuestro modelo predictivo, ya que su influencia es significativa en comparación con otras variables.

Upon analyzing the correlation matrix, it is evident that the column contributing the most to predicting the price of a laptop is the "Storage Capacity" column.
```python
# Este código importa varias clases y funciones de la biblioteca scikit-learn, que es ampliamente utilizada para el aprendizaje automático en Python.
# This code imports several classes and functions from the scikit-learn library, which is widely used for machine learning in Python.

# - LinearRegression: Esta clase se utiliza para crear un modelo de regresión lineal, que permite predecir un valor continuo basado en uno o más predictores.
# - LinearRegression: This class is used to create a linear regression model that predicts a continuous value based on one or more predictors.

# - train_test_split: Esta función se utiliza para dividir un conjunto de datos en dos partes: un conjunto de entrenamiento y un conjunto de prueba, lo que facilita la evaluación del rendimiento del modelo.
# - train_test_split: This function is used to split a dataset into two parts: a training set and a testing set, facilitating the evaluation of the model's performance.

# - mean_squared_error y r2_score: Estas son métricas utilizadas para evaluar la calidad del modelo. mean_squared_error calcula el error cuadrático medio entre las predicciones y los valores reales, mientras que r2_score mide qué tan bien las variables independientes explican la variabilidad de la variable dependiente.
# - mean_squared_error and r2_score: These are metrics used to assess the quality of the model. mean_squared_error calculates the mean squared error between predictions and actual values, while r2_score measures how well the independent variables explain the variability of the dependent variable.

# - make_classification y make_regression: Estas funciones generan conjuntos de datos sintéticos para problemas de clasificación y regresión, respectivamente, lo cual es útil para probar algoritmos y modelos.
# - make_classification and make_regression: These functions generate synthetic datasets for classification and regression problems, respectively, which is useful for testing algorithms and models.

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import make_classification
from sklearn.datasets import make_regression
```

```python
# Este código importa funciones necesarias de scikit-learn, luego separa las características (X) y la variable objetivo (y) del DataFrame df, eliminando la columna 'Price' para X.
# This code imports necessary functions from scikit-learn, then separates the features (X) and the target variable (y) from the DataFrame df by dropping the 'Price' column for X.
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Price'])
y = df['Price']
```

```python
# Esta línea divide los datos en conjuntos de entrenamiento y prueba, asignando el 70% de los datos a entrenamiento y el 30% a prueba, utilizando una semilla aleatoria para la reproducibilidad.
# This line splits the data into training and testing sets, allocating 70% of the data for training and 30% for testing, using a random seed for reproducibility.
X_train, X_test, Y_train, Y_test = train_test_split(X, y, train_size=0.7, random_state=1)
```

```python
# Esta línea crea un modelo de regresión lineal que incluye el término de intercepción y luego ajusta el modelo a los datos de entrenamiento (X_train, y_train).
# This line creates a linear regression model that includes the intercept term and then fits the model to the training data (X_train, y_train).
model = LinearRegression(fit_intercept=True)
model.fit(X_train, Y_train)
```

```python
# Esta línea imprime los coeficientes del modelo de regresión lineal, que representan la relación entre cada característica y la variable objetivo.
# This line prints the coefficients of the linear regression model, which represent the relationship between each feature and the target variable.
print("Coeficientes: ", model.coef_)
print("Ordenada al origen", model.intercept_)
```

```python
# Esta línea utiliza el modelo ajustado para predecir los valores de la variable objetivo en el conjunto de prueba (X_test).
# This line uses the fitted model to predict the target variable values in the test set (X_test).
y_pred = model.predict(X_test)
```

```python
# This line calculates the mean squared error (MSE) between the predicted values (y_pred) and the actual values (Y_test), which measures the average squared difference.
# Esta línea calcula el error cuadrático medio (MSE) entre los valores predichos (y_pred) y los valores reales (Y_test), que mide la diferencia cuadrática promedio.
mse = mean_squared_error(y_pred, Y_test)
print("MSE: ", mse)
```

```python
# This line calculates the R-squared score (r2), which indicates the proportion of variance in the dependent variable that can be explained by the independent variables.
# Esta línea calcula el coeficiente de determinación (r2), que indica la proporción de la varianza en la variable dependiente que puede ser explicada por las variables independientes.
r2 = r2_score(y_pred, Y_test)
print("r2", r2)
```

```python
# This line initializes a RandomForestRegressor with specified parameters, fits it to the training data (X_train, Y_train), and then predicts the target values for the test set (X_test) while also calculating the model's score on the test data.
# Esta línea inicializa un RandomForestRegressor con parámetros específicos, lo ajusta a los datos de entrenamiento (X_train, Y_train) y luego predice los valores objetivo para el conjunto de prueba (X_test), mientras también calcula la puntuación del modelo en los datos de prueba.
rfc = RandomForestRegressor(n_estimators=34, max_depth=15, random_state=42, min_samples_split=25)
rfc.fit(X_train, Y_train)
print(rfc.predict(X_test))
rfc.score(X_test, Y_test)
```
